```python
import csv
import os
labels = []
grs = []
temps = []
bpms = []
for file in os.listdir("."):
    if file[:4] == "euge":
        with open(file) as data:
            spamreader = csv.reader(data, delimiter=',', quotechar='|')
            i = 0
            for row in spamreader:
                if len(grs) - 1 < i:
                    grs.append([float(row[0])])
                    temps.append([float(row[1])])
                    bpms.append([float(row[2])])
                else:
                    grs[i].append(float(row[0]))
                    temps[i].append(float(row[1]))
                    bpms[i].append(float(row[2]))
                i+=1
            labels.append(file)
```

```python
import pandas as pd

df1 = pd.DataFrame(grs)
df2 = pd.DataFrame(temps)
df3 = pd.DataFrame(bpms)
df1.columns = labels
df2.columns = labels
df3.columns = labels

%matplotlib inline
df1.plot()
df2.plot()
df3.plot()
```
