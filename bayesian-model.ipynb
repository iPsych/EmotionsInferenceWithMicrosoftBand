## Resumen

En este trabajo se intenta desarrollar un modelo bayesiano que permita inferir cambios emocionales básicos a partir de datos fisiológicos. El sistema nervioso simpático es un sistema autónomo del ser humano que cumple entre otros papeles el de preparar al cuerpo para situaciones de acción, la clásica reacción de *fight or flight*. Cuando este sistema detecta una situación peligrosa envía señales a distintas partes del cuerpo, entre otras las glándulas sudoripadas, el corazón y los músculos del aparato locomotor. Realizando un modelo forward intentaremos modelar estos cambios entre estados de relajación y estados de acción, o sea intentaremos formalizar lo que ocurre cuando el sistema simpatico pasa a estar activo. Una vez logrado esto aplicaremos inferencia bayesiana para detectar dado un conjunto de mediciones fisiológicas el momento exacto de activación del sistema.

## Carga de Datos

```python

import csv
import os

def loadData(files):
    grs = []
    temps = []
    bpms = []
    for file in files:
        with open(file) as data:
            gr = []
            temp = []
            bpm = []
            spamreader = csv.reader(data, delimiter=',', quotechar='|')
            for row in spamreader:
                gr.append(float(row[0]))
                temp.append(float(row[1]))
                bpm.append(float(row[2]))
            grs.append(gr)
            temps.append(temp)
            bpms.append(bpm)
    return grs, temps, bpms

files = ["Data/record1.csv"]
GSRs, BTs, BPMs = loadData(files)
```

## El Modelo Bayesiano

El modelo obtenido está compuesto por una variable uniforme $\tau$ la cual representa el momento de cambio de estado del sistema nervioso, obviamente pueden darse múltiples cambios pero por simplicidad y ya que se amolda al dominio estudiado asumimos que hay un solo cambio de estado. Para nuestro caso particular representa el momento en que se pasa de un estado normal de relajación y tranquilidad a uno de mayor atención y tensión debido a los sobresaltos generados por el video.

$$ \tau \sim Uniform(0,N\_SAMPLES)$$

Dependiendo del valor de tau se decidirá qué variables uniformes utilizar para modelar la media de las distribuciones normales que representan los datos censados.

$$
\mu_{bpm} \sim \begin{cases}
             Normal_{initBPM}  & \text{if } \tau < timestamp \\
             Normal_{finBPM}  & \text{if } timestamp \ge \tau
       \end{cases} \quad
\mu_{gsr} \sim \begin{cases}
             Normal_{initGSR}  & \text{if } \tau < timestamp \\
             Normal_{finGSR}  & \text{if } timestamp \ge \tau
       \end{cases} \quad$$

$$\mu_{bt} \sim \begin{cases}
             Normal_{initBT}  & \text{if } \tau < timestamp \\
             Normal_{finBT}  & \text{if } timestamp \ge \tau
       \end{cases} \quad
$$

Como se vió en clase para los desvíos estándar de una gaussiana lo mejor es utilizar una distribución Gamma Inversa.

$$ \sigma_{bpm} \sim InverseGamma(3,6)$$
$$ \sigma_{gsr} \sim InverseGamma(31,31)$$
$$ \sigma_{bt} \sim InverseGamma(1,1)$$

Para escoger los parametros $\beta$ y $\alpha$ de la distribución Gamma Inversa se realiza el siguiente cálculo a partir del desvío y media deseados.

$$\mu = \frac{\beta}{(\alpha-1)}$$
$$\sigma = \frac{\beta^2}{\alpha(\alpha-1)^{2}}$$

Las distribuciones finales quedan definidas de la siguiente manera.

$$ N_{bpm} \sim Normal(\mu_{bpm},\sigma_{bpm})$$
$$ N_{gsr} \sim Normal(\mu_{gsr},\sigma_{gsr})$$
$$ N_{bt} \sim Normal(\mu_{bt},\sigma_{bt})$$

Los valores a priori de estas distribuciones fueron escogidos en base a la experiencia y literatura científica. Los pulsos por minuto suelen estar entre los 60 y 70 en estado de reposo, la temperatura en una persona sana es de 36°C, por último la conductancia cutánea puede variar mucho dependiendo la persona aunque suele rondar entre los 0.1 y los 10 $\mu$S (10000 y 100 K$\Omega$) debido a esta gran variabilidad se escogió el valor medio de los datos tomados como media inicial.


```python
import pymc3 as pm
import theano.tensor as tt
import numpy as np

exp_index = 0

n_samples = len(BPMs[exp_index])-1
model = pm.Model()

with model:
    initialMeanBPM = pm.Normal('Initial Mean BPM', mu=65, sd=3)
    finalMeanBPM = pm.Normal('Final Mean BPM', mu=65, sd=3)
    initialMeanGSR = pm.Normal('Initial Mean GSR', mu=7000, sd=30)
    finalMeanGSR = pm.Normal('Final Mean GSR', mu=7000, sd=30)
    initialMeanBT = pm.Normal('Initial Mean Body Temperature', mu=36, sd=0.5)
    finalMeanBT = pm.Normal('Final Mean Body Temperature', mu=36, sd=0.5)

    tau = pm.DiscreteUniform("tau", lower=0, upper=n_samples)
    
    idx = np.arange(n_samples+1)
    meanBPM = pm.math.switch(tau > idx, initialMeanBPM, finalMeanBPM)
    meanGSR = pm.math.switch(tau > idx, initialMeanGSR, finalMeanGSR)
    meanBT = pm.math.switch(tau > idx, initialMeanBT, finalMeanBT)

    sdBPM = pm.InverseGamma('SD BPM', alpha=3, beta=6)
    sdGSR = pm.InverseGamma('SD GSR', alpha=31, beta=31)
    sdBT = pm.InverseGamma('SD Body Temperature', alpha=1, beta=1)

    points1 = pm.Normal('BPM',
                       mu=meanBPM,
                       sd=sdBPM,
                       observed=BPMs[exp_index])

    points2 = pm.Normal('GSR',
                       mu=meanGSR,
                       sd=sdGSR,
                       observed=GSRs[exp_index])

    points3 = pm.Normal('Body Temperature',
                       mu=meanBT,
                       sd=sdBT,
                       observed=BTs[exp_index])

# fit model
with model:
    step = pm.Metropolis()
    tr = pm.sample(10000, step=step)
    
#pm.plots.traceplot(tr, ['SD BPM','SD GSR','SD Body Temperature','Initial Mean BPM', 'Final Mean BPM', 'Initial Mean GSR', 'Final Mean GSR', 'Initial Mean Body Temperature', 'Final Mean Body Temperature', 'tau']);
pm.plot_posterior(tr, ['Initial Mean BPM', 'Final Mean BPM', 'Initial Mean GSR', 'Final Mean GSR', 'Initial Mean Body Temperature', 'Final Mean Body Temperature', 'tau']);
```

## Resultados
Como se puede observar en los graficos obtenidos hay un punto en el que se da un gran cambio, donde las respuestas fisiológicas comienzan a ser explicadas mejor con una distribución totalmente distinta, este punto según los resultados obtenidos se da a los 4 minutos aproximadamente (263 segundos). Entre el estado inicial (Desde el comienzo hasta los primeros 4 minutos) y el final (De los 4 minutos hasta el final) se ven patrones completamente distintos las pulsaciones y la temperatura corporal son mayores en el estado final y la conductividad baja indicando que las glandulas sudoripadas estan funcionando.
